### Use the LabelEncoder & OneHotEncoder to work on categorical features
https://towardsdatascience.com/encoding-categorical-features-21a2651a065c

```python
# Categorical boolean mask
categorical_feature_mask = X_input_var.dtypes == object
categorical_feature_mask

# filter categorical columns using mask and turn it into a list
categorical_cols = X_input_var.columns[categorical_feature_mask].tolist()
categorical_cols
# One Hot Encoder
# from sklearn.preprocessing import OneHotEncoder
# ohe = OneHotEncoder(categorical_features = )
```

```python
# LabelEncoder converts each class under specificed feature to a numerical value.

# Instantiate a LabelEncoder object:

# Import labelencoder
from sklearn.preprocessing import LabelEncoder

# instantiate labelencoder object
le = LabelEncoder()

# Apply the LabelEncoder on each of the categorical columns:
X_input_var[categorical_cols] = X_input_var[categorical_cols].apply(lambda col: le.fit_transform(col))

X_input_var[categorical_cols].head(10)
```

Our categorical features are multi class, which makes it not a good choice for LabelEncoder since it bring a natural ordering for different classes. This can lead to certain inputs appearing to be 'greater' than others and poor performance.

Therefore, for our dataframe with multi class features, we will use OneHotEncoder as an additional step.

```python
# import OneHotEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.compose import ColumnTransformer
# prepare the categorical columns for OneHotEncoder
categorical_ix = X_input_var.select_dtypes(include=['object','bool']).columns
categorical_ix
```

```python
# define the data preparation for the columns
t = [('cat', OneHotEncoder(), categorical_ix)]
col_transform = ColumnTransformer(transformers=t)
```

```python
from numpy import mean
from numpy import std
from numpy import absolute

from sklearn.svm import SVR
from sklearn.pipeline import Pipeline
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python

# define the SVR model
model = SVR(kernel='rbf', gamma = 'scale', C=100)

# define the data preparation and modeling pipeline
pipeline = Pipeline(steps=[('prep', col_transform), ('m',model)])
```

```python
# define the model cross-validation configuration
cv = KFold(n_splits=10, shuffle=True, random_state=1)
# evaluate the pipeline using cross validation and calculate MAE
scores= cross_val_score(pipeline, X, y, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)

# summarize the model performance
print('MAE: %.3f (%.3f)' % (mean(scores),std(scores)))
```
